```python
board = [
    [7,8,0,4,0,0,1,2,0],
    [6,0,0,0,7,5,0,0,9],
    [0,0,0,6,0,1,0,7,8],
    [0,0,7,0,4,0,2,6,0],
    [0,0,1,0,5,0,9,3,0],
    [9,0,4,0,6,0,0,0,5],
    [0,7,0,3,0,0,0,1,2],
    [1,2,0,0,0,7,4,0,0],
    [0,4,9,2,0,6,0,0,7]
]


# 7 8 0  | 4 0 0  | 1 2 0
# 6 0 0  | 0 7 5  | 0 0 9
# 0 0 0  | 6 0 1  | 0 7 8
# - - - - - - - - - - - - - 
# 0 0 7  | 0 4 0  | 2 6 0
# 0 0 1  | 0 5 0  | 9 3 0
# 9 0 4  | 0 6 0  | 0 0 5
# - - - - - - - - - - - - - 
# 0 7 0  | 3 0 0  | 0 1 2
# 1 2 0  | 0 0 7  | 4 0 0
# 0 4 9  | 2 0 6  | 0 0 7


def solve(bo):                      # Backtracking algorithm
    
    find = find_empty(bo)           # 先找出"0"的element
    
    # base case
    if not find:                    # 都沒有"0"的element => 已完成
        
        return True
    
    else:
        
        row, col = find             # 此"0"的position
    
    # recursive call
    for i in range(1,10):               # 此i代表是欲填入的number
        
        if valid(bo, i, (row, col)):    # 如果此欲填入的number經valid()檢驗合格 => 填入number (填入i)
                                        # 如果此欲填入的number經valid()檢驗不合格 => 換一個i，繼續檢驗
            bo[row][col] = i            

            if solve(bo):               # 開始recursively call solve() => 填好此格後，繼續往下填下一個 "0"的element
                                        # 如果填下一格成功，就會一直持續到end
                return True             # 如果填下一格不成功 (在下一格中用盡所有的i，還是不行) => 代表上一格所填的number有問題 => backtracking ng

            bo[row][col] = 0            # 回到上一格 => 把上一格設回 "0" => 在上一格繼續try不同的i => 然後重覆above steps

    return False


def valid(bo, num, pos):                           # 判定欲填入的element是否match requirements
    
    # Check column
    for i in range(len(bo[0])):                    # 先看column
        
        if bo[pos[0]][i] == num and pos[1] != i:   # row是pos[0] => 看每一個i (看每一column) (horizontal)
                                                   # 當 pos[1] != i AND   bo[pos[0]][i] == num => 有重覆 => return False
            return False                           # i == pos[1]代表就是欲填入element的position，所以不須看此position => pos[1] != i

    # Check row
    for i in range(len(bo)):                       # 再看row
        
        if bo[i][pos[1]] == num and pos[0] != i:   # column是pos[1] => 看每一個i (看每一row) (vertical)
                                                   # 當 pos[0] != i  AND   bo[i][pos[1]] == num => 有重覆 => return False
            return False                           # i == pos[0]代表就是欲填入element的position，所以不須看此position => pos[0] != i

    # Check box  
    box_x = pos[1] // 3                            # 再看subgrids 
    box_y = pos[0] // 3                            # // 3 => 標示出此element所在的subgrid是屬於9宮格中的哪一個位置
        
    for i in range(box_y*3, box_y*3 + 3):          # 代表此subgrid的row的範圍 => 上面用// 3標示出9宮格 => 現在*3，標示出position
        
        for j in range(box_x * 3, box_x*3 + 3):    # 代表此subgrid的column的範圍 => 上面用// 3標示出9宮格 => 現在*3，標示出position
            
            if bo[i][j] == num and (i,j) != pos:   # 當 (i,j) != pos  AND   bo[i][pos[1]] == num => 有重覆 => return False
                
                return False                       # (i,j) == pos代表就是欲填入element的position，所以不須看此position => (i,j) != pos

    return True                                    # 經以上判定後，都沒有重複 => 代表此格欲填入的element matches requirements


def print_board(bo):                               # print出易讀的grid
    
    for i in range(len(bo)):                       
        
        if i % 3 == 0 and i != 0:
            
            print("- - - - - - - - - - - - - ")    # add seperations for every 3 rows

        for j in range(len(bo[0])):
            
            if j % 3 == 0 and j != 0:              # add seperations for every 3 columns
                
                print(" | ", end="")

                
            if j == 8:                             # print出 last elemnt in every row  & 換行
                
                print(bo[i][j])
                
            else:                                  # print出每一element  &  加" "  &  不換行
                
                print(str(bo[i][j]) + " ", end="")


def find_empty(bo):                            # find element "0" 的position 
    
    for i in range(len(bo)):
        
        for j in range(len(bo[0])):
            
            if bo[i][j] == 0:
                
                return (i, j)  # row, col      # return tuple (i, j)

    return None

print_board(board)
solve(board)
print("_________________________")
print_board(board)
```
